```python
# Import Dependencies
import os
import csv
import pandas as pd
import numpy as np
from functools import reduce
import matplotlib.pyplot as plt
from scipy.stats import linregress
import scipy.stats as st
```

```python
tasks_file = "../Resources/MCMG0007L2_CR_26Weeks_All_Claims_List.xls"
claims_df = pd.read_excel(tasks_file)
claims_df.head(20)
```

```python
claims_df=claims_df.drop([0,1,2,3,4,5,])
claims_df.head(10)
```

```python
claims_df.columns=claims_df.iloc[0]
claims_df.head()
```

```python
claims_df=claims_df.drop([6])
claims_df.head(10)
```

```python
claims_df=claims_df.drop(columns=["\n\n\n\nTeam", "\n\n\n\nOfficer", "\n\n\nWorker \nSurname", "\n\n\nWorker\nName", 
                                  "\n\nMedcert\nTo Date\nLatest", "\nLast \nCompensation\nPaid \nDate", "\n\n\nQBE \nFlag",
                                  "\nProjected \nDerived \nWeeks \nTo 30/06/2022", "\n\n\nNumerator\nIndicator", "\n\n\nDerived\nWeeks",
                                 "\n\nCompensation\nPaid to \nDate", "Total \nPre-injury\nWeekly \nShift \nAmount", 
                                  "\n\nProjected\n26 Weeks \nDate", "\n\n\nClaim \nStatus", "\n\n\n\nWEN", "\n\n\nIneligibility\nStatus",
                                 "\nEmployer\nExcess \nDay \nCount", "\n\nAgent \nReceived\nDate", "\n\n\nEmployer\nName",
                                 "\n\nAffliction\nNature \nCode", "\n\n\nIneligibility\nType", "\n\n\n\nTermination", 
                                  "\nResumed \nWork \nType \nIndicator"
                                 ])
claims_df.head()
                                  
```

```python
claims_df = claims_df.rename(columns={"\n\n\n\nClaim":"Claim", "\n\n\n\nAge":"Age", "\n\n\n\nGender":"Gender", 
                                      "\n\n\nInjury\nDate":"Injury_Date", "\n\nCeased\nWork \nDate":"Ceased_Work_Date",
                                      "\n\nResumed\nWork \nDate":"RTW_Date", "\n\nEmployer \nSegment \nDesciption":"Employer_Type", 
                                      "\n\nMedcert\nType \nLatest":"Certification"            
                                    })
claims_df.head()
```

```python
claims_df[["Claim", "Age"]] = claims_df[["Claim", "Age"]].apply(pd.to_numeric)
claims_df.dtypes
```

```python
rtw26 = "../Resources/RTW16.csv"
RTW26_df = pd.read_csv(rtw26)
RTW26_df.head()
```

```python
all_claims_df = claims_df.merge(RTW26_df, how='inner', on="Claim")
all_claims_df
```

```python
all_claims_df[["Injury_Date", "Ceased_Work_Date", "RTW_Date", "ARD"]]=all_claims_df[["Injury_Date", "Ceased_Work_Date", "RTW_Date", "ARD"]].apply(pd.to_datetime)
all_claims_df.dtypes
```

```python
all_claims_df["Lodgement_Gap"] = (all_claims_df["ARD"]-all_claims_df["Injury_Date"])/pd.Timedelta(days=1)
```

```python
all_claims_df["RTW_Duration"] = (all_claims_df["RTW_Date"]-all_claims_df["Ceased_Work_Date"])/pd.Timedelta(days=1)
all_claims_df.head()
```

```python
all_claims_df=all_claims_df.drop(columns=["\n\nAffliction \nNature \nDescription", "Total \nPre-injury\nWeekly \nOvertime \nAmount", "\nShift \nor \nOvertime \nIndicator",
                                         "Status", "Measure_Date_+3weeks", "\nWeekly Comp\nPayment \nEnd \nDate"])
all_claims_df
```

```python
current_rtw_df = all_claims_df.loc[(RTW26_df["Working"]=='Y')]
current_rtw_df = current_rtw_df.dropna(subset=['Status+3weeks'])

injury_code_rtw_group = current_rtw_df.groupby(["Injury_Code"])
injury_code_rtw = injury_code_rtw_group["Working"].count()
injury_code_rtw_df = pd.DataFrame({"RTW":injury_code_rtw}).reset_index()

current_nortw_df = all_claims_df.loc[(RTW26_df["Working"]=='N')]
current_nortw_df = current_nortw_df.dropna(subset=['Status+3weeks'])

injury_code_nortw_group = current_nortw_df.groupby(["Injury_Code"])
injury_code_nortw = injury_code_nortw_group["Working"].count()
injury_code_nortw_df = pd.DataFrame({"No_RTW":injury_code_nortw}).reset_index()
```

```python
merged_injury_df = injury_code_rtw_df.merge(injury_code_nortw_df, how="outer", on="Injury_Code")
merged_injury_df.head()
```

```python
merged_injury_df["RTW_Pct"] = round(merged_injury_df["RTW"]/(merged_injury_df["RTW"]+merged_injury_df["No_RTW"])*100,2)


merged_injury = merged_injury_df.loc[merged_injury_df['RTW']=='Na', 'RTW_Pct'].fillna(100)
    
merged_injury_df = merged_injury_df.sort_values(by="RTW_Pct", ascending=False)
merged_injury_df.head(64)
```
