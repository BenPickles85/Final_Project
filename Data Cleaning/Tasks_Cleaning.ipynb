```python
# Import Dependencies
import os
import csv
import pandas as pd
import numpy as np
```

```python
# Import CSV, read into pandas and display head
tasks_file = "../Resources/Novus_Daily_Tasks.xls"
overdue_tasks_df = pd.read_excel(tasks_file, sheet_name="Overdue_List")
overdue_tasks_df.head(20)
```

```python
overdue_tasks_df = overdue_tasks_df.drop([0,1,2,3,4,5])
overdue_tasks_df.head()
```

```python
overdue_tasks_df.columns=overdue_tasks_df.iloc[0]
```

```python
overdue_tasks_df=overdue_tasks_df.drop([6])
overdue_tasks_df.head()
```

```python
overdue_tasks_df.shape
```

```python
overdue_tasks_df = overdue_tasks_df.rename(columns={"\n\n\nTeam":"Team", "\nTask\nRole\nName":"Role_Name",
                                                    "\n\nTask\nName":"Task_Name", "\n\n\nClaim":"Claim", 
                                                    "\n\nCreation\nDate":"Creation_Date", "\n\nDue \nDate":"Due_Date",
                                                    "\n\nDays \nOverdue":"Days_Overdue"})
overdue_tasks_df.head()
```

```python
str_col = ["Team", "Role_Name", "Task_Name", "Claim", "Creation_Date", "Due_Date", "Days_Overdue"]
overdue_tasks_df=overdue_tasks_df[str_col]
overdue_tasks_df.head()
```

```python
overdue_tasks_df.to_csv('../Resources/OverdueTasks.csv')
```
