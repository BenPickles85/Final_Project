```python
# Import Dependencies
import os
import csv
import pandas as pd
import numpy as np
from functools import reduce
```

```python
# Import CSV, read into pandas and display head
tasks_file = "../Resources/EE - MRTW0026A - RTW Index - Back at Work 26 Weeks - Details and Stats.xlsx"
RTW26_df = pd.read_excel(tasks_file, sheet_name="26 weeks")
RTW26_df.head(20)
```

```python
RTW26_df=RTW26_df.drop([0,1,2,3,4,5,6])
RTW26_df.head(10)
```

```python
RTW26_df.columns=RTW26_df.iloc[0]
RTW26_df.head()
```

```python
RTW26_df=RTW26_df.drop([7])
RTW26_df.head()
```

```python
RTW26_df=RTW26_df.drop(columns=["\n\n\nWorker\nName", "\n\n\n\nTeam", "\n\n\nCase\nManager", "\n\n\nDeath\nDate",
                                "\n\nSuspension\nStart\nDate", "\n\nPayment\nEnd\nDate", "Resumed Work\nLast\nStatus\nUpdate\nDate",
                                "\n\n+3Wks\nFull Payment(3)\nFlag", "Weekly\nCompensation\nPaid\nFrom\nDate", "\n\n\nTermination\nFlag",
                                "\n\n\nSuspension\nFlag", "Latest\nWork Status\nEffective\nStart\nDate", "\n\n\nWIC\nCode", "Full\nWeekly\nPayment\nType 1\nFlag",
                                "Part\nWeekly\nPayment\nType 2\nFlag", "Full\nWeekly\nPayment\nType 3\nFlag", "\n\n+3Wks\nFull Payment(1)\nFlag", 
                                "\n\n+3Wks\nPart Payment(2)\nFlag", "\nClaim\nClassification\nChanged\nDate", "Wrk/Sts\nEffective\nfrom\nDate\nLatest",
                                "\n\nSuspenstion\nEnd\nDate", "\n\n\nResumed\nDate", "\n\nTermination\nStart\nDate", "\n\nTermination\nEnd\nDate",
                                "\n\nFirst\nCeased\nDate", "\n\nRTW\nDate\nLatest"
                               ])
RTW26_df.head()
```

```python
RTW26_df = RTW26_df.rename(columns={"\n\n\nClaim\nID":"Claim", "\n\n\nEmployer\nName":"Employer_Name", "\n\n\nMDM\nTeam":"Team",
                                    "\n\nAgent\nReceived\nDate":"ARD", "\n\n\nCompensation\nDays":"Days_Compensation",
                                    "\n\nClaim\nLodged\nDate":"Claim_Lodged_Date", "\n\n\nInjury\nCode":"Injury_Code",
                                    "\nPhysical\nor\nMental\nInjury":"Physical_or_Mental_Injury", 
                                    "\n\nMeasure\nDate\n@ 26 Weeks":"Measure_Date", "\n\nRTW\nType\nLatest":"RTW_Type",
                                    "\n\n\nWorking\nFlag":"Working", "\n\n\nEmployer\nNumber":"Employer_Number",
                                    "\n\nStatus\nExplanation\n@ 26 Weeks":"Description", "\n\n\nStatus\n@ 26 Weeks":"Status@26", 
                                    "\n\nStatus\n@ 26 Weeks\n+3 Weeks":"Status26plus"
                                    })
RTW26_df.head()
```

```python
RTW26_df.to_csv('../Resources/RTW26.csv', index=False)
```

```python
rtw_df = RTW26_df.loc[(RTW26_df["Working"]=='Y')]

MIrtw_df = rtw_df.loc[(rtw_df["Physical_or_Mental_Injury"]=="Mental")]
MIrtw_df = MIrtw_df.dropna(subset = ['Status26plus'])
teamMIrtw = MIrtw_df.groupby(["Team"])
teamMIrtwCount = teamMIrtw["Status26plus"].count()
teamMIrtw_df = pd.DataFrame({"MI_rtw": teamMIrtwCount}).reset_index()

physicalRTW_df = rtw_df.loc[(rtw_df["Physical_or_Mental_Injury"]=="Physical")]
physicalRTW_df = physicalRTW_df.dropna(subset = ['Status26plus'])
teamPhysical_rtw = physicalRTW_df.groupby(["Team"])
teamPhysicalrtwCount = teamPhysical_rtw["Status26plus"].count()
teamPhysicalrtw_df = pd.DataFrame({"Physical_rtw": teamPhysicalrtwCount}).reset_index()

noRTW_df = RTW26_df.loc[(RTW26_df["Working"]=='N')]

MInortw_df = noRTW_df.loc[(noRTW_df["Physical_or_Mental_Injury"]=="Mental")]
MInortw_df = MInortw_df.dropna(subset = ['Status26plus'])
teamMInortw = MInortw_df.groupby(["Team"])
teamMInortwCount = teamMInortw["Status26plus"].count()
teamMInortw_df = pd.DataFrame({"MI_nortw": teamMInortwCount}).reset_index()

physical_nortw_df = noRTW_df.loc[(noRTW_df["Physical_or_Mental_Injury"]=="Physical")]
physical_nortw_df = physical_nortw_df.dropna(subset = ['Status26plus'])
teamPhysical_nortw = physical_nortw_df.groupby(["Team"])
teamPhysicalnortwCount = teamPhysical_nortw["Status26plus"].count()
teamPhysicalnortw_df = pd.DataFrame({"Physical_nortw": teamPhysicalnortwCount}).reset_index()

teamPhysicalnortw_df
```

```python
data_frames = [teamMIrtw_df, teamPhysicalrtw_df, teamMInortw_df, teamPhysicalnortw_df]

merged_df = reduce(lambda  left,right: pd.merge(left,right,on=['Team'],
                                            how='outer'), data_frames)
merged_df
```

```python
merged_df["MI_RTW"] = round(merged_df["MI_rtw"]/(merged_df["MI_rtw"]+merged_df["MI_nortw"])*100,2)
merged_df["Phys_RTW"] = round(merged_df["Physical_rtw"]/(merged_df["Physical_rtw"]+merged_df["Physical_nortw"])*100,2)
merged_df["Total_RTW"] = round((merged_df["Physical_rtw"]+merged_df["MI_rtw"])
                                  /(merged_df["Physical_rtw"]+merged_df["Physical_nortw"]+merged_df["MI_rtw"]+merged_df["MI_nortw"])
                                 *100,2)

merged_df
```

```python
merged_df.to_csv('../Resources/team_RTW26.csv', index=False)
```
