```python
# Import Dependencies
import os
import csv
import pandas as pd
import numpy as np
```

```python
# Import CSV, read into pandas and display head
tasks_file = "../Resources/DRAP0028a - Invoices within Claims Teams and RAP.xlsx"
invoices_df = pd.read_excel(tasks_file, sheet_name="Other Queues")
invoices_df.head(20)
```

```python
invoices_df=invoices_df.drop([0,1,2,3,4,5])
invoices_df.head(10)
```

```python
invoices_df.columns=invoices_df.iloc[0]
invoices_df.head()
```

```python
invoices_df=invoices_df.drop([6])
invoices_df.head()
```

```python
invoices_df=invoices_df.drop(columns=["User ID", "Worker", "Due Date Changed", "Original Due Date"])
invoices_df.head()
```

```python
invoices_df = invoices_df.rename(columns={"Assigned To":"Assigned_To", "Task Type":"Task_Type", "Days Category":"Days_Category"      
                                    })
```

```python
invoices_df = invoices_df.dropna(subset = ["Team"])
invoices_df.head(50)
```

```python
invoices_df = invoices_df.reset_index(drop=True)
invoices_df.head()
```

```python
invoices_df.to_csv('../Resources/invoices.csv', index=False)
```
