```python
import numpy as np
import os
import csv
import pandas as pd
```

```python
rtw_df = pd.read_csv("26_week_B@W.csv")
rtw_df.head()
```

```python
rtw_df = rtw_df.rename(columns={"Employer Name":"Employer", "\n\n\nMDM\nTeam":"MDM Team", "\n\nAgent\nReceived\nDate":"ARD", 
                               "\n\nStatus\n@ 26 Weeks\n+3 Weeks":"RTW Status", "\nPhysical\nor\nMental\nInjury":"Injury Type",
                               "Weekly\nCompensation\nPaid\nFrom\nDate":"Comp Paid"})
rtw_df.head()
```

```python
small_df = rtw_df[["Employer", "MDM Team", "ARD", "RTW Status", "Injury Type", "Comp Paid"]]
small_df.head()
```

```python
MI_list_df = small_df.loc[(small_df["Injury Type"]=="Mental")]
MI_list = MI_list_df.groupby(["Employer", "MDM Team"])
MI_count_emp = MI_list["Injury Type"].count()

MI_df = pd.DataFrame({"MI Claims Count": MI_count_emp})
MI_df.sort_values("MI Claims Count", ascending=False).head()
```

```python
rtw_list_df = MI_list_df.loc[(small_df["RTW Status"]=="RTW")]
rtw_list = rtw_list_df.groupby(["Employer", "MDM Team"])
rtw_count_emp = rtw_list["RTW Status"].count()

rtw_count_df = pd.DataFrame({"RTW Count": rtw_count_emp}).reset_index()
rtw_count_df.head()
```

```python
nortw_list_df = MI_list_df.loc[(small_df["RTW Status"]=="No RTW")]
nortw_list = nortw_list_df.groupby(["Employer", "MDM Team"])
nortw_count_emp = nortw_list["RTW Status"].count()

no_rtw_count_df = pd.DataFrame({"No RTW Count": nortw_count_emp}).reset_index()
no_rtw_count_df.head()
```

```python
merged_df = rtw_count_df.merge(no_rtw_count_df, on=["Employer","MDM Team"], how="outer").fillna(0)
merged_df["RTW Percentage"] = round(merged_df["RTW Count"] / (merged_df["RTW Count"] + merged_df["No RTW Count"]) * 100, 2)
merged_df
```

```python
final_df = MI_df.merge(merged_df, on=["Employer","MDM Team"], how = "inner").reset_index()
final_df.sort_values("MI Claims Count", ascending=False).head(30)
```

```python
team_list = MI_list_df.groupby(["MDM Team"])
team_MI_count = team_list["Injury Type"].count()

team_df = pd.DataFrame({"MI Claims Count": team_MI_count})
team_df.sort_values("MI Claims Count", ascending=False).head(9)
```

```python
team_rtw_df = small_df.loc[(small_df["RTW Status"]=="RTW")]
rtw_list = team_rtw_df.groupby(["MDM Team"])
rtw_count_team = rtw_list["RTW Status"].count()

rtw_team_df = pd.DataFrame({"RTW": rtw_count_team}).reset_index()
rtw_team_df
```

```python
team_nortw_df = small_df.loc[(small_df["RTW Status"]=="No RTW")]
nortw_list = team_nortw_df.groupby(["MDM Team"])
nortw_count_team = nortw_list["RTW Status"].count()

nortw_team_df = pd.DataFrame({"No RTW": nortw_count_team}).reset_index()
nortw_team_df
```

```python
team_merged_df = rtw_team_df.merge(nortw_team_df, on=["MDM Team"], how="outer")
team_merged_df["RTW Percentage"] = round(team_merged_df["RTW"] / (team_merged_df["RTW"] + team_merged_df["No RTW"]) * 100, 2)
team_merged_df
```
